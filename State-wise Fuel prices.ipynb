## State-wise Fuel prices(EDA)

```python
import pandas as pd 
import matplotlib.pyplot as plt 
import numpy as np
import seaborn as sns
import os 
import warnings
```

```python
data = pd.read_csv("C:\\Users\\Nandhavena Navignan\\Downloads\\Project\\states_data_car.csv")
```

```python
sns.countplot(y=data["subsidy"],data=data)
```

```python
plt.figure(figsize=(10,8))
sns.set_theme(style='darkgrid')
sns.barplot(data["subsidy"],data["capital"],data=data)
plt.xlabel("Subisdy")
plt.ylabel("Capital")
plt.title("capital wise subisdy for vehicles")
plt.show()
```

```python
plt.figure(figsize=(15,10))
sns.scatterplot(data["diesel"],data["state"],data=data)
plt.xlabel("diesel")
plt.ylabel("state")
plt.title("diesel rate as per the state")
plt.show()
```

```python
plt.figure(figsize=(15,10))
sns.scatterplot(data["petrol"],data["state"],data=data)
plt.xlabel("petrol")
plt.ylabel("state")
plt.title("petrol as per the state")
plt.show()
```
