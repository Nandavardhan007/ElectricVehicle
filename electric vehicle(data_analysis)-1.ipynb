```python
plt.figure(figsize=(10,5))
sns.countplot(x='BodyStyle', data=ev_data, hue='PowerTrain')
plt.title('Body style count')
plt.xlabel('Body Style')
plt.ylabel('Count')
plt.show()
```

```python
#RANGE OF THE VEHICLE
data_1= ev_data.loc[ev_data['INR'] <=4500000]
data_2 = ev_data.loc[ev_data['INR'] >4500000]
price_1 = ['Less than INR(4500000)']
price_2 = ['More than INR(4500000)']
def range(dataframe, price):
    plt.figure(figsize=(20,5))
    sns.set_theme(style="darkgrid")
    sns.barplot('Model', 'Range_Km', data=ev_data, hue=ev_data['PowerTrain'])
    plt.title('''Range(Km) of EV'''.format(price))
    plt.ylabel('Range(Km)')
    plt.xlabel('vehicle model')
    plt.xticks(rotation = 90)
    plt.show()
range(data_1, price_1)
range(data_2, price_2)
    
```

```python
#range of the ev's vehicle
#price of the ev's vehicle
def range_price(dataframe, text):
    fig = plt.figure(figsize=(20, 5))
    a1 = plt.subplot()
    a1.bar(ev_data['CarName'], ev_data['Range_Km'], label='Range (Km)', color='yellow')
    plt.legend(loc='upper left', bbox_to_anchor = (0, 1.1))
    a2= a1.twinx()
    a2.scatter(ev_data['CarName'], ev_data['INR'], label = 'Price', color = 'green')
    plt.title('''RANGE (Km) vs PRICE(INR) OF EV's COSTING {}'''.format(text), fontsize=16)
    a1.set_xlabel('vehicel Models', size=20)
    a1.set_ylabel('Range (Km)', color = 'red')
    a2.set_ylabel('Price(INR)', color= 'black')
    plt.legend(loc = 'upper left', bbox_to_anchor = (0,1.1))
    a1.set_xticklabels(ev_data['CarName'], rotation = 'vertical')
    plt.show()
    
range_price(data_1, price_1)
range_price(data_2, price_2)
```

```python
#range-batterypack
def battery_range(dataframe, text):
    fig = plt.figure(figsize=(20,5))
    a1= plt.subplot()
    a1.bar(ev_data["CarName"], ev_data["Range_Km"], label='Range (Km)', color='blue')
    plt.legend(loc= "upper right", bbox_to_anchor=(0,1.105))
    a2 = a1.twinx()
    a2.scatter(ev_data["CarName"], ev_data["Battery_Pack Kwh"], label= "Battery Pack", color= 'orange')
    plt.title('''RANGE(Km) vs BATTERY PACK CAPACITY (kwh) of EV's costing {}'''.format(text), fontsize=12)
    a1.set_xlabel('vehicle Models', size= 20)
    a1.set_ylabel('Range (Km)', color = 'blue')
    a2.set_ylabel('Battery Pack Capacity (Kwh)', color='black')
    plt.legend(loc='upper left', bbox_to_anchor=(0,1))
    a1.set_xticklabels(ev_data['CarName'], rotation = 'vertical')
    plt.show()
    
battery_range(data_1, price_1)
battery_range(data_2, price_2)
```

```python
#Accelaration of ev_vehicle
def acc(dataframe, text):
    plt.figure(figsize=(20,5))
    sns.set_theme(style="whitegrid")
    sns.barplot('CarName', 'AccelSec', data=ev_data, hue=ev_data['PowerTrain'])
    plt.title('''Acceleration of EV's costing {}'''.format(text), fontsize=16)
    plt.ylabel('Acceleration (Seconds)')
    plt.xlabel('Model')
    plt.xticks(rotation = 90)
    plt.show()
    
acc(data_1, price_1)
acc(data_2, price_2)  
```
